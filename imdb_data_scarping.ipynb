# IMDB MOVIES SCRAPING

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.imdb.com/chart/top')
page
```

```python
Soup=BeautifulSoup(page.content)
Soup
```

```python
scraped_movies=[]
for i in Soup.find_all('td',class_=["titleColumn"]):
    scraped_movies.append(i.text.replace('\n',' '))
scraped_movies
```

```python
print(scraped_movies[:100])
```

```python
rating=[]
for i in Soup.find_all(class_=["ratingColumn imdbRating"]):
    rating.append(i.text.replace('\n','',))
rating
```

```python
print(rating[:100])
```

```python
year_of_release=[]
for i in Soup.find_all(class_=["secondaryInfo"]):
    year_of_release.append(i.text.replace(',',' '))
year_of_release
```

```python
print(year_of_release[0:100])
```

```python
print(len(scraped_movies),len(rating),len(year_of_release))
```

# Making Dataframe

```python
import pandas as pd
df=pd.DataFrame({'scraped_movies':scraped_movies,'rating':rating,'year_of_release':year_of_release})
df
```

# Top rated Indian movies

```python
page=requests.get('https://www.imdb.com/list/ls009997493/?sort=list_order,asc&st_dt=&mode=detail&page=1')
page
```

```python
Soup=BeautifulSoup(page.content)
Soup
```

```python
movies=[]
for i in Soup.find_all('h3',class_=["lister-item-header"]):
    movies.append(i.text.replace('\n',' '))
movies
```

```python
rating=[]
for i in Soup.find_all('div',class_="ipl-rating-star small"):
    rating.append(i.text.replace('\n',''))
                                  
rating
```

```python
year_of_release=[]
for i in Soup.find_all('span',class_=["lister-item-year text-muted unbold"]):
    year_of_release.append(i.text.replace('(I) ',' '))
year_of_release
```

```python
print(len(movies),len(rating),len(year_of_release))
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'movies':movies,'rating':rating,'year_of_release':year_of_release})
df
```
